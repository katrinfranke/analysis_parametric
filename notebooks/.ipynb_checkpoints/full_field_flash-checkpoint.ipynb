## Full-field flash – exploratory analysis

This notebook explores how model neurons respond to a **full-field luminance flash**
that steps from a uniform gray background to a brighter level.

Three things are investigated:

1. **Stimulus construction** – build a 12-frame achromatic flash stimulus
   (gray pre-flash context, then bright flash).
2. **Contrast sweep** – vary flash contrast from 0 to 1 and plot predicted
   responses for a handful of example neurons.
3. **Gaze sweep** – fix contrast at a high value and shift the simulated gaze
   position; show how the predicted responses of the same neurons change.

### Stimulus layout

| Frames | Content | Valid predictions? |
|--------|---------|-------------------|
| 0–2    | gray (mean_lum = 0.5) | no – skipped (skip_samples = 3) |
| 3–11   | bright flash | **yes** – 9 valid predictions |

The first `skip_samples = 3` frames are used as temporal context only.
All 9 valid predictions therefore fall inside the flash period.

### Flash luminance
```
flash_lum = mean_lum + contrast * (1 - mean_lum)
```
With `mean_lum = 0.5`:
- contrast = 0 → flash_lum = 0.5 (no flash, pure gray)
- contrast = 0.5 → flash_lum = 0.75
- contrast = 1.0 → flash_lum = 1.0 (maximum white)

## 1 – Full-field flash stimulus

The stimulus is a (T, 3, H, W) float32 array in [0, 1].
- **Pre-flash frames (0–2):** uniform gray at `mean_lum`.
- **Flash frames (3–11):** uniform bright at `mean_lum + contrast * (1 − mean_lum)`.

All three RGB channels are set to the same value (achromatic / luminance flash).

````python
def make_flash_frames(
    contrast: float = 1.0,
    mean_lum: float = 0.5,
    num_frames: int = 12,
    onset_frame: int = 3,
    H: int = 236,
    W: int = 420,
) -> np.ndarray:
    """
    Build an achromatic full-field flash stimulus.

    Args:
        contrast:    Flash contrast in [0, 1].
                     0 = no flash (all gray), 1 = maximum white.
        mean_lum:    Background (pre-flash) luminance in [0, 1].
        num_frames:  Total number of frames (model input length).
        onset_frame: Frame index at which the flash turns on.
        H, W:        Spatial dimensions matching the model input.

    Returns:
        frames: (num_frames, 3, H, W) float32 in [0, 1].
    """
    flash_lum = float(mean_lum) + float(contrast) * (1.0 - float(mean_lum))
    flash_lum = float(np.clip(flash_lum, 0.0, 1.0))

    frames = np.full((num_frames, 3, H, W), mean_lum, dtype=np.float32)
    frames[onset_frame:] = flash_lum
    return frames
````

````python
# Preview: show luminance of each frame for two contrasts
fig, axes = plt.subplots(1, 2, figsize=(14, 3))

for ax, contrast in zip(axes, [0.5, 1.0]):
    frames = make_flash_frames(contrast=contrast)
    lum_per_frame = frames[:, 0, 0, 0]  # all pixels identical → just read one

    ax.step(np.arange(len(lum_per_frame)), lum_per_frame, where='post',
            color='steelblue', linewidth=2)
    ax.axvspan(0, 3, alpha=0.08, color='gray', label='skipped (context)')
    ax.axvspan(3, 12, alpha=0.08, color='gold', label='valid predictions')
    ax.set_ylim(-0.05, 1.1)
    ax.set_xlabel('Frame')
    ax.set_ylabel('Luminance')
    ax.set_title(f'Full-field flash  (contrast = {contrast})')
    ax.legend(fontsize=8)
    sns.despine(ax=ax)

plt.tight_layout()
plt.show()
````

## 2 – Contrast sweep

We present the flash at several contrast levels and record the mean and
time-resolved predicted response of each example neuron.

**Prediction strategy:**
- One 12-frame window per contrast level.
- `wrapper.predict()` returns shape `(U, T_valid)` = `(653, 9)`.
- We average over the 9 valid time steps to get one scalar per neuron.

````python
CONTRASTS = np.array(np.linspace(-1, 1, 20))
MEAN_LUM = 0.5

# contrast_responses[i, u, t] = predicted response of neuron u at time t for contrasts[i]
contrast_responses = []  # list of (U, T_valid) arrays

for c in CONTRASTS:
    frames = make_flash_frames(contrast=c, mean_lum=MEAN_LUM)
    pred, _ = wrapper.predict(frames)   # (U, T_valid)
    contrast_responses.append(pred)

contrast_responses = np.stack(contrast_responses, axis=0)  # (n_contrasts, U, T_valid)
print(f"contrast_responses shape: {contrast_responses.shape}  → (n_contrasts, U, T_valid)")
````

````python
unit_idx = np.random.choice(indices_v3a)

fig, ax = plt.subplots(figsize=(6, 4))
cmap = plt.cm.RdBu_r  # diverging colormap suits -1 to 1 range
norm = plt.Normalize(-1, 1)

T_valid = contrast_responses.shape[2]
time_axis = np.arange(skip_samples, skip_samples + T_valid) / 30.0  # seconds, 30 fps

for ci, (c, color) in enumerate(zip(CONTRASTS, cmap(norm(CONTRASTS)))):
    ax.plot(time_axis, contrast_responses[ci, unit_idx, :],
            color=color, label=f'{c:.1f}', linewidth=1.5)

ax.set_title(f'Neuron {unit_idx}', fontsize=10)
ax.set_xlabel('Time after flash  (s)', fontsize=8)
ax.set_ylabel('Predicted response', fontsize=8)
sns.despine(ax=ax)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, label='Flash contrast', shrink=0.8, pad=0.02)
fig.suptitle('Time-resolved response to full-field flash – contrast sweep', fontsize=13)
plt.tight_layout()
plt.show()
````

## 3 – Gaze sweep

The model receives gaze position as a `(T, 2)` float32 array in the same
coordinate system as the training data (eye-tracker coordinates in pixels).

- **`gaze_txy = np.zeros((T, 2))`** – fixation at the screen centre.
- **`gaze_txy[:, 0] = x_offset`** – shift gaze horizontally by `x_offset` pixels.
- **`gaze_txy[:, 1] = y_offset`** – shift gaze vertically by `y_offset` pixels.

The screen is 420 × 236 px (≈ 63 × 35 deg at 6.7 px/deg), so offsets of
±100 px correspond to ≈ ±15 deg — a large but plausible saccade.

We fix contrast at **1.0** and sweep gaze along the horizontal axis first,
then sweep 2-D positions on a coarser grid.

> **Note:** The gaze coordinate system depends on the preprocessing pipeline.
> If the training data normalises gaze to a different range, adjust the offsets below.

````python
# --- Contrast sweep with saccade-like 2D gaze trajectory ---
# Saccade parameters
GAZE_X_START, GAZE_X_END = 0,  100.0   # px
GAZE_Y_START, GAZE_Y_END =  0,   30.0   # px  (adjust as needed)
SACCADE_ONSET_FRAME = 4                       # frame on which saccade begins
SACCADE_DURATION_FRAMES = 3                   # typical ~100 ms at 30 Hz ≈ 3 frames

def make_saccade_trajectory(
    x_start, x_end, y_start, y_end,
    num_frames, onset_frame, duration_frames
):
    """
    Smooth saccade using a symmetric velocity profile (raised cosine / Hann window).
    Pre-saccade: fixation at start position.
    Saccade: position ramps with a rise-and-fall velocity profile.
    Post-saccade: fixation at end position.
    """
    txy = np.zeros((num_frames, 2), dtype=np.float32)

    # Cumulative displacement weights from a Hann (raised-cosine) velocity profile
    t_sac = np.arange(duration_frames)
    vel   = 0.5 * (1 - np.cos(2 * np.pi * t_sac / duration_frames))  # Hann window
    cum   = np.cumsum(vel)
    cum   = cum / cum[-1]   # normalise so final value == 1.0

    for frame in range(num_frames):
        if frame < onset_frame:
            txy[frame] = [x_start, y_start]
        elif frame < onset_frame + duration_frames:
            w = cum[frame - onset_frame]
            txy[frame, 0] = x_start + w * (x_end - x_start)
            txy[frame, 1] = y_start + w * (y_end - y_start)
        else:
            txy[frame] = [x_end, y_end]

    return txy

gaze_txy_saccade = make_saccade_trajectory(
    GAZE_X_START, GAZE_X_END,
    GAZE_Y_START, GAZE_Y_END,
    NUM_FRAMES, SACCADE_ONSET_FRAME, SACCADE_DURATION_FRAMES
)

# --- Run contrast sweep with saccade trajectory ---
gaze_saccade_responses = []
for c in CONTRASTS:
    frames = make_flash_frames(contrast=c, mean_lum=MEAN_LUM,
                               num_frames=NUM_FRAMES, onset_frame=ONSET_FRAME)
    pred, _ = wrapper.predict(frames, gaze_txy=gaze_txy_saccade)   # (U, T_valid)
    gaze_saccade_responses.append(pred)
gaze_saccade_responses = np.stack(gaze_saccade_responses, axis=0)
# shape: (n_contrasts, U, T_valid)

print(f"gaze_saccade_responses shape: {gaze_saccade_responses.shape}")
print(f"gaze_x trajectory (px): {gaze_txy_saccade[:, 0]}")
print(f"gaze_y trajectory (px): {gaze_txy_saccade[:, 1]}")

# --- Quick sanity-check plot of the trajectory itself ---
frame_axis = np.arange(NUM_FRAMES)
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

axes[0].plot(frame_axis, gaze_txy_saccade[:, 0], label='x', color='steelblue')
axes[0].plot(frame_axis, gaze_txy_saccade[:, 1], label='y', color='tomato')
axes[0].axvspan(SACCADE_ONSET_FRAME,
                SACCADE_ONSET_FRAME + SACCADE_DURATION_FRAMES,
                alpha=0.15, color='gray', label='saccade window')
axes[0].set_xlabel('Frame'); axes[0].set_ylabel('Gaze position (px)')
axes[0].set_title('Gaze position over time'); axes[0].legend(fontsize=8)

axes[1].plot(gaze_txy_saccade[:, 0], gaze_txy_saccade[:, 1],
             'o-', color='purple', markersize=4)
axes[1].plot(*gaze_txy_saccade[0],  'g^', markersize=8, label='start')
axes[1].plot(*gaze_txy_saccade[-1], 'rs', markersize=8, label='end')
axes[1].set_xlabel('Gaze x (px)'); axes[1].set_ylabel('Gaze y (px)')
axes[1].set_title('2D saccade trajectory'); axes[1].legend(fontsize=8)

sns.despine()
plt.tight_layout()
plt.show()
````

````python
# Time-resolved contrast sweep – saccade gaze condition
T_valid   = gaze_saccade_responses.shape[2]
time_axis = np.arange(skip_samples, skip_samples + T_valid) / 30.0
frame_axis = np.arange(NUM_FRAMES) / 30.0   # full gaze trajectory in seconds

cmap = plt.cm.RdBu_r
norm = plt.Normalize(-1, 1)
unit_idx = np.random.choice(indices_v3a)

fig, axes = plt.subplots(
    2, 1, figsize=(6, 6),
    gridspec_kw=dict(height_ratios=[3, 1], hspace=0.08),
    sharex=False   # time axes have same range but different tick density is fine
)
ax_resp, ax_gaze = axes

# --- Top: neural responses ---
for ci, c in enumerate(CONTRASTS):
    ax_resp.plot(time_axis, gaze_saccade_responses[ci, unit_idx, :],
                 color=cmap(norm(c)), linewidth=1.5)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax_resp, label='Flash contrast', shrink=0.8, pad=0.02)

ax_resp.set_ylabel('Predicted response', fontsize=8)
ax_resp.set_title(
    f'Neuron {unit_idx}  |  saccade {GAZE_X_START:+.0f}→{GAZE_X_END:+.0f} px (x),'
    f'  {GAZE_Y_START:+.0f}→{GAZE_Y_END:+.0f} px (y)',
    fontsize=9
)
ax_resp.tick_params(labelbottom=False)   # shared x, hide bottom ticks on top panel
sns.despine(ax=ax_resp, bottom=True)

# --- Bottom: gaze x and y position ---
ax_gaze.plot(frame_axis, gaze_txy_saccade[:, 0],
             color='steelblue', linewidth=1.5, label='x')
ax_gaze.plot(frame_axis, gaze_txy_saccade[:, 1],
             color='tomato',    linewidth=1.5, label='y')

# shade the saccade window
sac_t_start = SACCADE_ONSET_FRAME / 30.0
sac_t_end   = (SACCADE_ONSET_FRAME + SACCADE_DURATION_FRAMES) / 30.0
ax_gaze.axvspan(sac_t_start, sac_t_end, alpha=0.12, color='gray', label='saccade')

ax_gaze.set_xlabel('Time after flash (s)', fontsize=8)
ax_gaze.set_ylabel('Gaze pos. (px)', fontsize=8)
ax_gaze.legend(fontsize=7, loc='upper left', frameon=False)
sns.despine(ax=ax_gaze)

# Align x limits across both panels
x_min = min(time_axis[0], frame_axis[0])
x_max = max(time_axis[-1], frame_axis[-1])
ax_resp.set_xlim(x_min, x_max)
ax_gaze.set_xlim(x_min, x_max)

fig.suptitle('Contrast sweep – saccade gaze trajectory', fontsize=12)
plt.tight_layout()
plt.show()
````

````python
# Average over high positive contrasts first → (n_neurons, T)
static_mean_c  = contrast_responses[0:3, :, :][:, indices_v3a, :].mean(axis=0)
saccade_mean_c = gaze_saccade_responses[0:3, :, :][:, indices_v3a, :].mean(axis=0)

from scipy import stats

n = len(indices_v3a)

# Mean and 95% CI across neurons → (T,)
static_mean  = static_mean_c.mean(axis=0)
saccade_mean = saccade_mean_c.mean(axis=0)

static_ci  = stats.sem(static_mean_c,  axis=0)
saccade_ci = stats.sem(saccade_mean_c, axis=0)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(time_axis, static_mean,  color='steelblue', linewidth=1.5, label='Static')
ax.fill_between(time_axis,
                static_mean - static_ci,
                static_mean + static_ci,
                color='steelblue', alpha=0.2)

ax.plot(time_axis, saccade_mean, color='tomato', linewidth=1.5, label='Saccade')
ax.fill_between(time_axis,
                saccade_mean - saccade_ci,
                saccade_mean + saccade_ci,
                color='tomato', alpha=0.2)

ax.set_xlabel('Time after flash (s)', fontsize=8)
ax.set_ylabel('Predicted response', fontsize=8)
ax.set_title(f'V3A  |  high positive contrasts  |  n={n} neurons  |  ±SEM', fontsize=10)
ax.legend(fontsize=8, frameon=False)
sns.despine(ax=ax)
plt.tight_layout()
plt.show()
````

````python
# ── Analysis window parameters (absolute frame indices into T_valid) ──────────
# Inspect your time_axis to set these; e.g. time_axis[i] gives frame i's time
# Baseline: pre-flash / pre-saccade period
BASELINE_FRAMES   = np.arange(0, 2)      # e.g. frames 0–2  (~0.00–0.07 s)
# Suppression: during saccade movement
SUPPRESS_FRAMES   = np.arange(3, 5)      # e.g. frames 4–6  (~0.13–0.20 s)
# Rebound: post-saccade window
REBOUND_FRAMES    = np.arange(6, 8)     # e.g. frames 7–10 (~0.23–0.33 s)
# ─────────────────────────────────────────────────────────────────────────────

def mean_in_window(responses, frames):
    """Mean over specified frame indices. responses: (n_contrasts, n_neurons, T)"""
    return responses[:, :, frames].mean(axis=2)   # → (n_contrasts, n_neurons)

# --- Baseline (same for both conditions, use static) ---
baseline    = mean_in_window(contrast_responses,      BASELINE_FRAMES)   # (C, U)

# --- Suppression window ---
static_supp  = mean_in_window(contrast_responses,      SUPPRESS_FRAMES)  # (C, U)
saccade_supp = mean_in_window(gaze_saccade_responses,  SUPPRESS_FRAMES)  # (C, U)

# --- Rebound window ---
static_reb   = mean_in_window(contrast_responses,      REBOUND_FRAMES)   # (C, U)
saccade_reb  = mean_in_window(gaze_saccade_responses,  REBOUND_FRAMES)   # (C, U)

# --- Suppression Index: SI = 1 - (saccade / static) during saccade window ---
# SI > 0 : suppressed during saccade
# SI < 0 : enhanced during saccade
# Clamp denominator to avoid division by ~0 for silent neurons
_eps = 1e-6
SI = 1.0 - (saccade_supp / (static_supp + _eps))          # (C, U)
SI = SI.T                                                   # → (U, C)

# --- Rebound Ratio: how much larger is post-saccadic response vs static ---
# RR > 1 : rebound enhancement
# RR < 1 : still suppressed post-saccade
rebound_ratio = saccade_reb / (static_reb + _eps)          # (C, U)
rebound_ratio = rebound_ratio.T                             # → (U, C)

# --- Rebound Excess: absolute difference (same units as predicted response) ---
rebound_excess = (saccade_reb - static_reb).T               # (U, C)

# --- Quick sanity check ---
print(f"SI            shape: {SI.shape}             "
      f"mean±std: {SI.mean():.3f} ± {SI.std():.3f}")
print(f"rebound_ratio shape: {rebound_ratio.shape}  "
      f"mean±std: {rebound_ratio.mean():.3f} ± {rebound_ratio.std():.3f}")
print(f"rebound_excess shape: {rebound_excess.shape} "
      f"mean±std: {rebound_excess.mean():.3f} ± {rebound_excess.std():.3f}")

# --- Print per-contrast summary (mean across neurons) ---
print("\nPer-contrast summary (mean across all neurons):")
print(f"{'Contrast':>10}  {'SI mean':>10}  {'SI std':>8}  "
      f"{'RR mean':>10}  {'RR std':>8}")
for ci, c in enumerate(CONTRASTS):
    print(f"{c:>10.2f}  {SI[:, ci].mean():>10.3f}  {SI[:, ci].std():>8.3f}  "
          f"{rebound_ratio[:, ci].mean():>10.3f}  {rebound_ratio[:, ci].std():>8.3f}")
````

````python
# --- 2-D gaze sweep ---
# gaze_2d_mean[i, j, u] = mean response of neuron u with gaze = (GAZE_X_VALUES_2D[j], GAZE_Y_VALUES[i])

# Use a coarser grid for the 2-D sweep to keep runtime reasonable
GAZE_X_2D = np.array([-150, -75, 0, 75, 150], dtype=np.float32)
GAZE_Y_2D = np.array([-75, 0, 75], dtype=np.float32)

gaze_2d_mean = np.zeros((len(GAZE_Y_2D), len(GAZE_X_2D), 653), dtype=np.float32)

for i, gy in enumerate(GAZE_Y_2D):
    for j, gx in enumerate(GAZE_X_2D):
        gaze_txy = np.zeros((NUM_FRAMES, 2), dtype=np.float32)
        gaze_txy[:, 0] = gx
        gaze_txy[:, 1] = gy
        pred, _ = wrapper.predict(flash_frames, gaze_txy=gaze_txy)   # (U, T_valid)
        gaze_2d_mean[i, j, :] = pred.mean(axis=1)                    # (U,)

print(f"gaze_2d_mean shape: {gaze_2d_mean.shape}  → (n_gy, n_gx, U)")
````
